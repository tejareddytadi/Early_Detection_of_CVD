```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score , classification_report
from sklearn import tree
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("/content/drive/MyDrive/machine_learning/cardio_train_kartik.csv", sep=";")
print("Number of rows: %d" % df.shape[0])
print("Number of columns: %d" % df.shape[1])
```

```python
print(df.head())
```

```python
df.drop(df[(df['height'] > df['height'].quantile(0.975)) | (df['height'] < df['height'].quantile(0.025))].index,inplace=True)
df.drop(df[(df['weight'] > df['weight'].quantile(0.975)) | (df['weight'] < df['weight'].quantile(0.025))].index,inplace=True)
df.drop(df[(df['ap_hi'] > df['ap_hi'].quantile(0.975)) | (df['ap_hi'] < df['ap_hi'].quantile(0.025))].index,inplace=True)
df.drop(df[(df['ap_lo'] > df['ap_lo'].quantile(0.975)) | (df['ap_lo'] < df['ap_lo'].quantile(0.025))].index,inplace=True)
len(df)
```

```python
print(df.isnull().sum())
```

```python
df['age'] = round(df['age']/365.25,2)
```

```python
df.drop(['id'], axis=1, inplace=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
X = df.drop(columns=["cardio"])
y = df["cardio"]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model6 = tree.DecisionTreeClassifier(max_depth = 10 , min_samples_leaf = 1 , min_samples_split = 5)

model6.fit(X_train , y_train)
```

```python
y_pred6 = model6.predict(X_test)
```

```python
print(f"Accuracy for DecisionTree: {accuracy_score(y_test , y_pred6)}")
class_report = classification_report(y_test, y_pred6)
print("Classification Report:\n", class_report)
conf_matrix = confusion_matrix(y_test, y_pred6)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
#filling NaN values with respect to the analysis


data2['education'] = data2['education'].fillna(data2['education'].mean())
data2['cigsPerDay'] = data2['cigsPerDay'].fillna(data2['cigsPerDay'].mean())
data2['BPMeds'] = data2['BPMeds'].fillna(data2['BPMeds'].mode()[0])
data2['totChol'] = data2['totChol'].fillna(data2['totChol'].mean())
data2['BMI'] = data2['BMI'].fillna(data2['BMI'].mean())
data2['heartRate'] = data2['heartRate'].fillna(data2['heartRate'].mean())
data2['glucose'] = data2['glucose'].fillna(data2['glucose'].mean())
```

```python
data2.isnull().sum()
```

```python
data2['sex'].replace(['F','M'],[0,1], inplace=True)
data2['is_smoking'].replace(['NO','YES'],[0,1], inplace=True)
```

```python
data2 = data2.drop(['id', 'education'], axis=1)
```

```python
# Calculate the correlation matrix
correlation_matrix = data2.corr()

# Select columns with correlation greater than or equal to 0.15 with respect to "TenYearCHD"
selected_columns = correlation_matrix["TenYearCHD"] >= 0.15

# Filter the correlation matrix based on selected columns
filtered_data = correlation_matrix[selected_columns]
filtered_data
```
